**Data Visualization**

```python
year = df1['HIRE_DATE'].unique()
years_count = df1['HIRE_DATE'].value_counts(dropna = False).sort_index()
plt.figure(figsize = (18,10))
sns.barplot(x = year, y = years_count, palette = "Paired")
plt.title('Details of HIRE DATE',fontsize=30)
plt.xticks(rotation = 50)
plt.xlabel('HIRE DATE',fontsize=30)
plt.ylabel('HIRE in Each Year',fontsize=30)
plt.show()
```

```python
plt.subplots(figsize=(20,10))
sns.barplot(df1['JOB_ID'].value_counts()[:10].index,df1['JOB_ID'].value_counts()[:10].values,palette='Set2')
plt.title('JOB ID',fontsize=30)
plt.xticks(rotation = 50)
plt.show()
```

```python
plt.subplots(figsize=(20,10))
sns.countplot(df1["MANAGER_ID"],order=df1['MANAGER_ID'].value_counts().index,palette="BuGn_r");
plt.xticks(rotation=50)
plt.xlabel("MANAGER ID", fontsize=30)
plt.ylabel("count",fontsize=30)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')
l = ['SH_CLERK', 'ST_CLERK','FI_ACCOUNT']
s = [50,50,50]
ax.pie(s, labels = l,autopct='%1.2f%%')
plt.show()
```
